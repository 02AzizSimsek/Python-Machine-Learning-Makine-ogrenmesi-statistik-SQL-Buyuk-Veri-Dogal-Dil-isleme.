```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill",y="tip",data=df);#dogrusal çizgi için lmplot kattık lm lineeri ifade eder
```

dogru etrafındaki mavi sapmayı ifade eder

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df);
```

sigara içemeyenler için ödeme arttıkca bahşiş artıyor fakat içenlerde bu şiddet azalmış egim daha az

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",data=df);
```

egimlerde fark var onu gördük ve ögüne göre iki grafik çıktı ,fakat buradaki korelasyon anlamlı mı onunda araştırmasını bilimsel yolla yapılması gerekiyor

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",row="sex",data=df);
```
