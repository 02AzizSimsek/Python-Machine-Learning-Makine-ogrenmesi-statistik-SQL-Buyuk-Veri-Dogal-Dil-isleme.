**SCATTERPLOT MATRİSİ**

sayısal degişkenler arasında ilişkiyi matris olarak bakmak 

```python
import seaborn as sns;
```

```python
#iris=>bu veri seti içinde çiçek türleri var bunlar üzerindeki özellikleri ifade eder
iris=sns.load_dataset("iris")
df=iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df);
```

scatterplot grafikler ayrık ise arada toz bulutu gibi bir yapı yoksa eger bu 

iki sayısal degişken arasında ilişki yok demektir 
birbirinden ayrı yerlerde kümelenme olmuşsa eger çaprazlama durumundan kaynaklamışır

```python
sns.pairplot(df,hue="species");
```

```python
#grafige dogru ekleme işlemi

sns.pairplot(df,kind="reg");
```

```python
sns.pairplot(df, kind="reg",hue="species");
```
