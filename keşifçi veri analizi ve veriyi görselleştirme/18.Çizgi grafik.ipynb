```python
import seaborn as sns
fmri=sns.load_dataset("fmri")
df=fmri.copy()
df.head()
```

```python
df["timepoint"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
sns.lineplot(x="timepoint",y="signal",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="event",data=df);
```

```python
sns.lineplot(x="timepoint",
             y="signal",
             hue="event",
             style="event",
             markers=True ,
             dashes=False, data=df);#ortalama noktaları işaretlemiş olduk 
```

```python
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             markers=True ,
             dashes=False, data=df);#grafik bize daha açıklanabilir hale gelmiş oldu
```
