**BÜYÜK SAYILAR YASASI-BİNOM DAGILIMI-İŞ UYGULAMASI(reklam harcamasaı Optimizasyonu)**

bir rassal degişkenin uzun kararlılığını tanımlayan olasılık teoremidir

```python
import numpy as np
rng=np.random.RandomState(123)
for i in np.arange(1,21):
    deney_sayisi=2**i#2 üzeri i demektir
    yazi_turalar=rng.randint(0,2,size=deney_sayisi)
    yazi_olasiliklari=np.mean(yazi_turalar)
    print("Atis Sayisi:",deney_sayisi,"---","Yazi Olasiligi:%.2f" % (yazi_olasiliklari))
```
