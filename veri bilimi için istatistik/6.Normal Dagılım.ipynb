normal dagıldıgı bilinen sürekli rassal degişkenler için olasılık heasplanması için kullanılır 




problem :bir yatırım/toplantı öncesinde gelecek ay ile ilgili satışların belirli degerlerde gerçekleşemesi olasıklları belirlenmek isteniyor 

detaylar :

-dagılımın normal oldugu biliniyor 

-aylık ortalama satış sayısı 80k,standart sapması 5k

-90k'dan fazla satış yapma olasılıgı nedir?








```python
from scipy.stats import norm
```

```python
#90'dan fazla olması 
```

```python
1-norm.cdf(90,80,5)#cdf =kümülatif yogunluk fonksiyonu,grafikte alan 1 den küçük diye 1 den çıkardık
```

```python
#70'dan fazla olması 
```

```python
1-norm.cdf(70,80,5)
```

```python
#73'daen az olması 
```

```python
norm.cdf(73,80,5)
```

```python
#85 ile 90 arasında olması 
```

```python
norm.cdf(90,80,5)-norm.cdf(85,80,5)
```

```python
1-belirsilik altında karar vermeye çalışıyoruz
2.uygun olasılık fonkisyonları ile olasılık hesapları yapıyoruz 
```
