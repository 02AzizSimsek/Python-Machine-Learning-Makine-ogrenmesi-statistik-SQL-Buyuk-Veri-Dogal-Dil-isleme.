```python
from scipy.stats import poisson
```

```python
lambda_=0.1
```

```python
rv=poisson(mu=lambda_)
print(rv.pmf(k=0))
```

```python
print(rv.pmf(k=3))
```

```python
print(rv.pmf(k=5))
```
