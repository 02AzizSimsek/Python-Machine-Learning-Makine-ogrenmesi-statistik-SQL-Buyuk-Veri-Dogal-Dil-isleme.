**Kesikli ve sürekli olasılık dagılımları**


bernoulli Dağılımı=>başarılı-başarısız,olumlu-olumsuz şeklinde iki sonuçlu olayalar ile ilgilendiginde kullanılan kesikli olasılık dagılımıdır.




**Bernoulli Dağılımı**

${\displaystyle f(k;p)=p^{k}(1-p)^{1-k}\quad{\text (for)}k\in\{0,1\}}$

${\displaystyle\operatorname{E}\left(x\right)=p}$

$\operatorname {Var}[X]=pq=p(1-p)$

```python
from scipy.stats import bernoulli
```

```python
 p=0.6#tura gelme olaslıgı 0.6 olarak biliniyor
```

```python
rv=bernoulli(p)
rv.pmf(k=0)#olasılık kitle fonksiyonu buraya giriliyor,K=1 girince tura k=0 girince yazı olma olasılıgı 
```
