```python
import numpy as np
import pandas as pd
V1=np.array([1,3,6,5,7])
V2=np.array([7,7,5,8,12])
V3=np.array([6,12,5,6,14])
df=pd.DataFrame(
{
    "V1":V1,
    "V2":V2,
    "V3":V3})
df=df.astype(float)
df
```

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)#bu işlemi bütün degişkenlere uyguladı
```

```python
df.mean()#dönüştürme yapılmamışların ortalaması 
```

## Normalizasyon

```python
preprocessing.normalize(df)
```

## 0-1 Dönüşümü

female-male kısmını 1 ve 0 dönüştürme işlemi yapıyoruz 

```python
from sklearn.preprocessing import LabelEncoder

lbe=LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])#veri setini dönüştürdü 
```

```python
df["yeni_sex"]=lbe.fit_transform(df["sex"])
```

```python
df#hangi kategorik degişkenin hangi degerine ne geldigini iyi anlamak gerekiyor 
```

# One-Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
import pandas as pd
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
```

```python
df_one_hot=pd.get_dummies(df, columns=["sex"],prefix=["sex"])#prefix yapılacak olan degişimlerin ön isimlendirimesinin ne olacagını ifade etmektir
```

```python
df_one_hot.head()
```

bir degişkeni ifade eden başka bir degişken varsa buna dummy degişken tuzagı denir

```python
pd.get_dummies(df,columns=["day"],prefix=["day"]).head()
```

one hot dönüşüm faydaları

1.genelde nümerik işlemler yaptıgımız için kullanırız 
2.kategorik degişkenler içindeki alt sınıfların etkilerini daha net hissederiz
3.birbirleri üzerinden oluşturulma durumunu ortadan kaldırır
