**Data Frame (join) birleştirme işlemleri**

```python
import numpy as np 
import pandas as pd 
m =np.random.randint(1,30, size =(5,3))
df1=pd.DataFrame(m,columns=["var1","var2","var3"])
df1
```

```python
df2=df1+99#bir data ve sayıyı toplayıp yeni bir dataframe elde ettik fakat iki tablo şeklinde degilde sayısal olarak artış sagladı
```

```python
df2
```

```python
pd.concat([df1,df2])#burada iki tane dataframe'i tablo olarak birleştirdi,fakat indexlerde hata var
```

```python
pd.concat([df1,df2],ignore_index=True)#önce fonksiyonun kendisini inceledik ve oradan ignore kısmını aldık ve çözüm için deneme yaptık 
```
